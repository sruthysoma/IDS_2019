```python
#IMPORTING ALL NECESSARY LIBRARIES

import pandas as pd
import matplotlib.pyplot as mplot
import plotly.express as px
import numpy as np
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
#LOADING THE DATASET USING PANDAS

dataset = pd.read_csv("Victims_of_Rape.csv")
dataset.head(20)
```

***
### DATA CLEANING

```python
#CHECKING THE NUMBER OF NULL ENTRIES IN EVERY COLUMN

dataset.isnull().sum()
```

***
The step below is to check for negative values in the dataset and replace them with zero:

```python
num = dataset._get_numeric_data()
num[num < 0] = 0
dataset.head(20)
```

***
Now, the missing values and values which aren't numerical (NaN) will be filled with mean value of their respective columns.

```python
dataset = dataset.fillna(dataset.mean())
dataset.head(20)
```

```python
#CHECKING IF ALL NULL VALUES HAVE BEEN ERADICATED

dataset.isnull().sum()
```

### NORMALIZATION OF THE DATA

```python
from sklearn import preprocessing as pre
x_array = np.array(dataset['Victims_Between_18-30_Yrs'])
normalized_X = pre.normalize([x_array])

# Distribution Plot (a.k.a. Histogram)
sns.distplot(normalized_X)
```

### CORRELATION MATRIX

Below is the transformation of categorical variables to numerical:

```python
from sklearn.preprocessing import LabelEncoder
X = dataset.iloc[:,:].values
labelencoder_X = LabelEncoder()
X[:,0] = labelencoder_X.fit_transform(X[:,0])
X[:,2] = labelencoder_X.fit_transform(X[:,2])
Y = pd.DataFrame(X)
Y.tail()
```

The heatmap representing the correlation matrix is as follows:

```python
sns.set(rc={'figure.figsize':(11.8,10)})
cor_data = dataset
cor_matrix = cor_data.corr()
sns.heatmap(cor_matrix,cmap="YlGnBu")
```

From the correlation matrix (heatmap) we can infer that the correlation between the number of rape cases reported and vitims between 18-30 years is the highest (represented by the darkest colour). This implies that the age group containing maximum number of victims of rape from 2001-2010 in India is 18-30 years.
